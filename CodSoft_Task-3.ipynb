Apply Elbow Technique

```python
krange=range(1,10)
ss=[]
for k in krange:
  kms=KMeans(n_clusters=k)
  kms.fit(df[['petal_length','petal_width']])
  ss.append(kms.inertia_)
```

```python
plt.xlabel('k_rng')
plt.ylabel("Sum of Squared errors")
plt.plot(krange,ss)
```

```python
ss
```

Applying KMean Alorithm

```python
kms=KMeans(n_clusters=3, random_state=0)
y_predicted= kms.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(100)
```

Measuring for Accuracy

```python
cms = confusion_matrix(df.species,df.cluster)
cms
```

```python
true_label = df.species
predicted_label= df.cluster

cms = confusion_matrix(true_label, predicted_label)
class_label = ['Setosa', 'versicolor', 'virginica']

plt.imshow(cms, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_mark = np.arange(len(class_label))
plt.xticks(tick_mark, class_label)
plt.yticks(tick_mark, class_label)

# Fill matrix with values
for i in range(len(class_label)):
    for j in range(len(class_label)):
        plt.text(j, i, str(cms[i][j]), ha='center', va='center', color='white')

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```
